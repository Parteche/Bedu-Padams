```python
#importamos pandas
import pandas as pd
#importamos numpys
import numpy as np
```

```python
#Importamos la base de datos CSV
df = pd.read_csv('/content/drive/MyDrive/Books_Data_Clean.csv')
```

```python
#Metodo para ver las 5 primeras filas de la base de datos
df.head()
```

```python
#Queremos saber que tipo de columnas tenemos:
df.dtypes
```

```python
#La columna Publishing Year tiene tipo de dato float64 es decir decimal y deberia ser int64 osea entero
#Con el metodo .astype() cambiamos el tipo de dato y con el metodo .fillna() logramos que no nos afecte
#los registros que tienen NA
df['Publishing Year'] = df['Publishing Year'].fillna(0).astype('int64')
#Checamos de nuevo nuestros tipos de datos para confirmar que si cambio el tipo de dato de la columna 'Publishing Year'
print(df.dtypes)
```

```python
#Checamos de nuevo los primeros 5 registros de nuestra base de datos
df.head()
#Ahora si la columna de 'Publishing Year' tiene un formato adecuado
```

Breve explicacion de lo que contiene cada columna:

Book name: El titulo del libro

Author: El nombre del autor

Criterios que tomaremos en cuenta para la limpieza de nuestros datos:

```python
#Datos faltantes
#Verificar valores nulos en todas las columnas:
print(df.isnull().sum())
#El resultado nos arrija que la columna Book Name tiene 23 registros nulos y la columna language_code tiene 53
```

```python
#Usaremos el metodo .info() para tener mas informacion sobre nuestra base de datos
df.info()
#Vemos que nuestra base de datos tiene 1070 filas y 15 columnas, por lo que valdria la pena eliminar los
#registros nulos, ya que tenemos suficientes datos para seguir trabajando
```

```python
#Eliminando registros nulos
df.dropna(inplace=True)
#Verificamos de nuevo registros nulos
print(df.isnull().sum())
#Los resultados son favorables, cada vez tenemos una base de datos mas limpia
```
